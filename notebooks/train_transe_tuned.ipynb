```python
import torch
from pathlib import Path
from pykeen.triples import TriplesFactory
from pykeen.pipeline import pipeline

PROJECT_ROOT = Path.home() / "xai-knowledge-graph"
TRIPLES_PATH = PROJECT_ROOT / "data" / "triples" / "xai_kg.tsv"
MODEL_DIR    = PROJECT_ROOT / "models" / "transe_tuned"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {DEVICE}")
print(f"Triples file:  {TRIPLES_PATH}")
```

```python
tf = TriplesFactory.from_path(str(TRIPLES_PATH))
print(f"Loaded {tf.num_triples:,} triples")
print(f"  Entities:  {tf.num_entities:,}")
print(f"  Relations: {tf.num_relations}")

# 80/10/10 split for training, validation, test
training, testing, validation = tf.split([0.8, 0.1, 0.1], random_state=42)
print(f"\nSplit sizes:")
print(f"  Train:      {training.num_triples:,}")
print(f"  Validation: {validation.num_triples:,}")
print(f"  Test:       {testing.num_triples:,}")
```

```python
result = pipeline(
    training=training,
    testing=testing,
    validation=validation,
    model="TransE",
    model_kwargs=dict(embedding_dim=128),       
    loss="NSSALoss",
    loss_kwargs=dict(margin=9.0, adversarial_temperature=1.0),
    training_kwargs=dict(num_epochs=200, batch_size=512),
    optimizer_kwargs=dict(lr=0.001),
    negative_sampler_kwargs=dict(num_negs_per_pos=64),
    evaluation_kwargs=dict(batch_size=256),
    device=DEVICE,
    random_seed=42,
)
print("\n✓ Training complete")
```

```python
metrics_df = result.metric_results.to_df()
print(metrics_df.to_string())

# Pull out the key headline metrics
key_metrics = result.metric_results.to_flat_dict()
print(f"\n=== TransE_tuned Test-Set Metrics ===")
print(f"  Hits@1:       {key_metrics.get('both.realistic.hits_at_1', 0):.4f}")
print(f"  Hits@3:       {key_metrics.get('both.realistic.hits_at_3', 0):.4f}")
print(f"  Hits@10:      {key_metrics.get('both.realistic.hits_at_10', 0):.4f}")
print(f"  Mean Rank:    {key_metrics.get('both.realistic.arithmetic_mean_rank', 0):.2f}")
print(f"  MRR:          {key_metrics.get('both.realistic.inverse_harmonic_mean_rank', 0):.4f}")
```

```python
MODEL_DIR.mkdir(parents=True, exist_ok=True)
result.save_to_directory(str(MODEL_DIR))
print(f"Model saved to {MODEL_DIR}")
print(f"\nContents:")
for p in sorted(MODEL_DIR.rglob("*")):
    if p.is_file():
        print(f"  {p.relative_to(MODEL_DIR)}  ({p.stat().st_size / 1024:.1f} KB)")
```
